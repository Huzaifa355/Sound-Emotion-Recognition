```python
import os
import json
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, mixed_precision
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Enable mixed precision for faster training
mixed_precision.set_global_policy('mixed_float16')

print(f"TensorFlow: {tf.__version__}")
print(f"GPU: {tf.config.list_physical_devices('GPU')}")

# ============================================================================
# MOUNT DRIVE & PATHS
# ============================================================================
from google.colab import drive
drive.mount('/content/drive')

DATASET_PATH = '/content/dataset/DataSet'

# ============================================================================
# HYPERPARAMETERS - 3-CHANNEL MEL-SPECTROGRAM (Mel + Delta + Delta-Delta)
# ============================================================================

# Audio parameters
SAMPLE_RATE = 16000
DURATION = 3.0
MAX_LENGTH = int(SAMPLE_RATE * DURATION)

# MEL-SPECTROGRAM parameters (3-channel physics-aware)
N_MELS = 128
N_FFT = 2048
HOP_LENGTH = 512
MAX_TIME_STEPS = 94  # ~3s audio with hop=512
N_CHANNELS = 3  # Mel + Delta + Delta-Delta

# Training parameters
BATCH_SIZE = 16
EPOCHS = 90
TARGET_LR = 1e-4  # Target learning rate for warmup
WARMUP_RATIO = 0.1  # First 10% of training for warmup
VALIDATION_SPLIT = 0.15
TEST_SPLIT = 0.10

# Model parameters
CNN_FILTERS = [64, 128, 256]
LSTM_UNITS = 128
DROPOUT_RATE = 0.4
L2_REG = 0.001
LABEL_SMOOTHING = 0.1

# Transfer Learning Control
USE_TRANSFER_LEARNING = True  # Set to True to load previous model
FREEZE_CNN_LAYERS = False  # Set to True to freeze CNN layers during transfer learning

# Seed
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)

# Auto-detect emotions
EMOTION_CLASSES = sorted([d for d in os.listdir(DATASET_PATH)
                         if os.path.isdir(os.path.join(DATASET_PATH, d))])
NUM_CLASSES = len(EMOTION_CLASSES)

print(f"\n✓ {NUM_CLASSES} emotions: {EMOTION_CLASSES}")
print(f"✓ Using 3-Channel Mel-spectrogram: {N_MELS} mel bands × {N_CHANNELS} channels")

# ============================================================================
# CHECKPOINT SETUP
# ============================================================================

checkpoint_dir = '/content/drive/MyDrive/ser_3chanel_checkpoints'
output_dir = '/content/drive/MyDrive/ser_3chanel_output'
os.makedirs(checkpoint_dir, exist_ok=True)
os.makedirs(output_dir, exist_ok=True)

print(f"\n✓ Checkpoints: {checkpoint_dir}")
print(f"✓ Outputs: {output_dir}")

# ============================================================================
# AUGMENTATION
# ============================================================================

def safe_augment_audio(audio):
    """Apply light augmentation (training only)"""
    aug_choice = np.random.randint(0, 5)

    if aug_choice == 0:
        # Add noise
        noise = np.random.randn(len(audio)) * 0.003
        return audio + noise
    elif aug_choice == 1:
        # Time shift
        shift = int(np.random.uniform(-0.1, 0.1) * len(audio))
        return np.roll(audio, shift)
    elif aug_choice == 2:
        # Pitch shift
        n_steps = np.random.uniform(-1.0, 1.0)
        return librosa.effects.pitch_shift(audio, sr=SAMPLE_RATE, n_steps=n_steps)
    elif aug_choice == 3:
        # Time stretch
        rate = np.random.uniform(0.95, 1.05)
        return librosa.effects.time_stretch(audio, rate=rate)

    return audio  # 20% no augmentation

# ============================================================================
# DATA LOADING
# ============================================================================

def get_file_paths_and_labels(dataset_path):
    """Collect file paths and labels"""
    file_paths, labels = [], []

    for emotion_idx, emotion in enumerate(EMOTION_CLASSES):
        emotion_dir = os.path.join(dataset_path, emotion)
        if not os.path.exists(emotion_dir):
            continue

        files = [f for f in os.listdir(emotion_dir)
                if f.endswith(('.wav', '.mp3', '.flac'))]

        for f in files:
            file_paths.append(os.path.join(emotion_dir, f))
            labels.append(emotion_idx)

        print(f"  {emotion}: {len(files)} files")

    return file_paths, labels

print("\nCollecting files...")
file_paths, labels = get_file_paths_and_labels(DATASET_PATH)
print(f"Total: {len(file_paths)} files")

# Train/Val/Test split
train_paths, test_paths, train_labels, test_labels = train_test_split(
    file_paths, labels, test_size=TEST_SPLIT, random_state=SEED, stratify=labels
)

train_paths, val_paths, train_labels, val_labels = train_test_split(
    train_paths, train_labels,
    test_size=VALIDATION_SPLIT / (1 - TEST_SPLIT),
    random_state=SEED, stratify=train_labels
)

print(f"Split: Train={len(train_paths)}, Val={len(val_paths)}, Test={len(test_paths)}")

# ============================================================================
# PREPROCESSING - 3-CHANNEL MEL-SPECTROGRAM (PILLAR 1: Physics-Aware)
# ============================================================================

def load_and_preprocess_audio_3channel(file_path, label, augment=False):
    """
    Load audio and extract 3-channel Mel-spectrogram:
    - Channel 1: Mel-Spectrogram (Log scale)
    - Channel 2: Delta (Velocity)
    - Channel 3: Delta-Delta (Acceleration)
    Output shape: (time_steps, n_mels, n_channels) = (94, 128, 3)
    """

    def _load(path, do_aug):
        path = path.numpy().decode('utf-8')
        do_aug = do_aug.numpy()

        try:
            # Load audio
            audio, sr = librosa.load(path, sr=SAMPLE_RATE, duration=DURATION)

            # Augmentation (training only)
            if do_aug:
                audio = safe_augment_audio(audio)

            # Pad/truncate
            if len(audio) < MAX_LENGTH:
                audio = np.pad(audio, (0, MAX_LENGTH - len(audio)))
            else:
                audio = audio[:MAX_LENGTH]

            # Extract Mel-spectrogram
            mel_spec = librosa.feature.melspectrogram(
                y=audio,
                sr=SAMPLE_RATE,
                n_mels=N_MELS,
                n_fft=N_FFT,
                hop_length=HOP_LENGTH,
                fmax=8000
            )

            # Convert to log scale (dB)
            mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

            # Transpose to (time, frequency)
            mel_spec_db = mel_spec_db.T  # (time, n_mels)

            # Compute Delta (Velocity)
            delta = librosa.feature.delta(mel_spec_db.T).T

            # Compute Delta-Delta (Acceleration)
            delta_delta = librosa.feature.delta(mel_spec_db.T, order=2).T

            # Stack channels: (time, n_mels, 3)
            mel_3channel = np.stack([mel_spec_db, delta, delta_delta], axis=-1)

            # Normalize each channel independently
            for i in range(3):
                channel = mel_3channel[:, :, i]
                mel_3channel[:, :, i] = (channel - np.mean(channel)) / (np.std(channel) + 1e-8)

            # Fixed time steps
            if mel_3channel.shape[0] < MAX_TIME_STEPS:
                pad_width = ((0, MAX_TIME_STEPS - mel_3channel.shape[0]), (0, 0), (0, 0))
                mel_3channel = np.pad(mel_3channel, pad_width)
            else:
                mel_3channel = mel_3channel[:MAX_TIME_STEPS, :, :]

            return mel_3channel.astype(np.float32)

        except Exception as e:
            print(f"Error: {path}: {e}")
            return np.zeros((MAX_TIME_STEPS, N_MELS, N_CHANNELS), dtype=np.float32)

    aug_flag = tf.constant(augment, dtype=tf.bool)
    mel_3ch = tf.py_function(_load, [file_path, aug_flag], tf.float32)
    mel_3ch.set_shape([MAX_TIME_STEPS, N_MELS, N_CHANNELS])  # (94, 128, 3)

    return mel_3ch, label

def create_dataset(paths, labels, batch_size, shuffle=True, augment=False):
    """Create tf.data pipeline"""
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))

    if shuffle:
        ds = ds.shuffle(1000, seed=SEED)

    ds = ds.map(
        lambda x, y: load_and_preprocess_audio_3channel(x, y, augment=augment),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    ds = ds.batch(batch_size).prefetch(3)
    return ds

# Create datasets
print("\nCreating 3-channel pipelines...")
train_ds = create_dataset(train_paths, train_labels, BATCH_SIZE, shuffle=True, augment=True)
val_ds = create_dataset(val_paths, val_labels, BATCH_SIZE, shuffle=False, augment=False)
test_ds = create_dataset(test_paths, test_labels, BATCH_SIZE, shuffle=False, augment=False)

print("✓ Train: augmentation=True")
print("✓ Val/Test: augmentation=False")
print(f"✓ Input shape: ({MAX_TIME_STEPS}, {N_MELS}, {N_CHANNELS})")

# ============================================================================
# PILLAR 2: ATTENTION MECHANISM (Custom Layer for Silence Handling)
# ============================================================================

class AttentionLayer(layers.Layer):
    """
    Custom Attention mechanism to handle silence/padding.
    Learns to weight LSTM outputs, allowing the model to ignore
    irrelevant timesteps (silence at start/end of clips).
    """
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        # input_shape: (batch, time_steps, features)
        self.W = self.add_weight(
            name='attention_weight',
            shape=(input_shape[-1], 1),
            initializer='glorot_uniform',
            trainable=True
        )
        self.b = self.add_weight(
            name='attention_bias',
            shape=(input_shape[1], 1),
            initializer='zeros',
            trainable=True
        )
        super(AttentionLayer, self).build(input_shape)

    def call(self, inputs):
        # inputs: (batch, time_steps, features)
        # Compute attention scores
        e = tf.nn.tanh(tf.matmul(inputs, self.W) + self.b)  # (batch, time_steps, 1)
        a = tf.nn.softmax(e, axis=1)  # Normalized attention weights

        # Apply attention weights
        output = inputs * a  # Weighted features
        output = tf.reduce_sum(output, axis=1)  # (batch, features)

        return output

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[-1])

# ============================================================================
# PILLAR 4: ADAPTIVE LEARNING - WarmUp + Cosine Decay
# ============================================================================

class WarmUpCosineDecay(keras.optimizers.schedules.LearningRateSchedule):
    """
    Custom Learning Rate Schedule:
    - Warmup: Linear increase from 0 to target_lr over warmup_steps
    - Cosine Decay: Smooth decay to near-zero after warmup
    """
    def __init__(self, target_lr, total_steps, warmup_steps):
        super(WarmUpCosineDecay, self).__init__()
        self.target_lr = target_lr
        self.total_steps = total_steps
        self.warmup_steps = warmup_steps
        self.decay_steps = total_steps - warmup_steps

    def __call__(self, step):
        # Cast step to float32 for calculations
        step = tf.cast(step, tf.float32)
        warmup_steps = tf.cast(self.warmup_steps, tf.float32)
        decay_steps = tf.cast(self.decay_steps, tf.float32)

        # Warmup phase
        warmup_lr = (self.target_lr / warmup_steps) * step

        # Cosine decay phase
        decay_step = step - warmup_steps
        cosine_decay = 0.5 * (1 + tf.cos(
            tf.constant(np.pi, dtype=tf.float32) * decay_step / decay_steps
        ))
        decay_lr = self.target_lr * cosine_decay

        # Return warmup or decay based on current step
        return tf.cond(
            step < warmup_steps,
            lambda: warmup_lr,
            lambda: decay_lr
        )

    def get_config(self):
        return {
            'target_lr': self.target_lr,
            'total_steps': self.total_steps,
            'warmup_steps': self.warmup_steps
        }

# ============================================================================
# PILLAR 3: MODEL ARCHITECTURE (Rectangular Kernels + Attention)
# ============================================================================

def build_adaptive_ser_model(input_shape, num_classes):
    """
    Advanced SER Model with:
    - 3-Channel Input (Mel + Delta + Delta-Delta)
    - Rectangular CNN Kernels (Time/Freq focus)
    - Bidirectional LSTM with Attention
    - Adaptive Learning Schedule
    """
    from tensorflow.keras.regularizers import l2

    inputs = layers.Input(shape=input_shape)  # (94, 128, 3)

    # CNN Block 1 - Time Focus (5x3 kernel)
    x = layers.Conv2D(CNN_FILTERS[0], kernel_size=(5, 3), padding='same',
                     kernel_regularizer=l2(L2_REG), name='cnn_time_focus')(inputs)
    x = layers.BatchNormalization(name='bn1')(x)
    x = layers.Activation('relu', name='relu1')(x)
    x = layers.MaxPooling2D((2, 2), name='pool1')(x)
    x = layers.Dropout(DROPOUT_RATE, name='dropout1')(x)

    # CNN Block 2 - Frequency Focus (3x5 kernel)
    x = layers.Conv2D(CNN_FILTERS[1], kernel_size=(3, 5), padding='same',
                     kernel_regularizer=l2(L2_REG), name='cnn_freq_focus')(x)
    x = layers.BatchNormalization(name='bn2')(x)
    x = layers.Activation('relu', name='relu2')(x)
    x = layers.MaxPooling2D((1, 2), name='pool2')(x)
    x = layers.Dropout(DROPOUT_RATE, name='dropout2')(x)

    # CNN Block 3 - Texture Focus (3x3 kernel)
    x = layers.Conv2D(CNN_FILTERS[2], kernel_size=(3, 3), padding='same',
                     kernel_regularizer=l2(L2_REG), name='cnn_texture_focus')(x)
    x = layers.BatchNormalization(name='bn3')(x)
    x = layers.Activation('relu', name='relu3')(x)
    x = layers.MaxPooling2D((1, 2), name='pool3')(x)
    x = layers.Dropout(DROPOUT_RATE, name='dropout3')(x)

    # Reshape for LSTM
    shape = x.shape
    x = layers.Reshape((shape[1], shape[2] * shape[3]), name='reshape_for_lstm')(x)

    # Bidirectional LSTM (return_sequences=True for Attention)
    x = layers.Bidirectional(
        layers.LSTM(LSTM_UNITS, return_sequences=True,
                   recurrent_dropout=0.1,
                   kernel_regularizer=l2(L2_REG)),
        name='bilstm'
    )(x)
    x = layers.BatchNormalization(name='lstm_bn')(x)

    # ATTENTION LAYER (handles silence/padding)
    x = AttentionLayer(name='attention')(x)
    x = layers.Dropout(DROPOUT_RATE, name='attention_dropout')(x)

    # Dense classifier
    x = layers.Dense(256, activation='relu', kernel_regularizer=l2(L2_REG), name='dense')(x)
    x = layers.Dropout(DROPOUT_RATE, name='dense_dropout')(x)

    # Output (float32 for mixed precision)
    outputs = layers.Dense(num_classes, activation='softmax', dtype='float32', name='output')(x)

    return keras.Model(inputs=inputs, outputs=outputs, name='Adaptive_SER_3Channel')

# ============================================================================
# RESUME TRAINING LOGIC - CHECK CSV LOG
# ============================================================================

def get_last_completed_epoch(log_file):
    """Read CSV log and return the last completed epoch number"""
    if not os.path.exists(log_file):
        print(f"⚠ No training log found at {log_file}")
        return 0

    try:
        df = pd.read_csv(log_file)
        if len(df) == 0:
            print("⚠ Training log is empty")
            return 0

        last_epoch = df['epoch'].iloc[-1]
        print(f"✓ Found training log with {len(df)} completed epochs")
        print(f"✓ Last completed epoch: {last_epoch}")
        return int(last_epoch) + 1  # +1 because epoch is 0-indexed

    except Exception as e:
        print(f"⚠ Error reading training log: {e}")
        return 0

# ============================================================================
# TRANSFER LEARNING LOGIC
# ============================================================================

def load_pretrained_model():
    """Try to load a pretrained model for transfer learning"""
    possible_paths = [
        os.path.join(checkpoint_dir, 'best_adaptive_model.keras'),
        os.path.join(output_dir, 'best_adaptive_model.keras'),
        os.path.join(output_dir, 'final_adaptive_model.keras'),
    ]

    for model_path in possible_paths:
        if os.path.exists(model_path):
            try:
                print(f"\n🔄 Loading pretrained model from: {model_path}")
                # Load with custom objects
                model = keras.models.load_model(
                    model_path,
                    custom_objects={
                        'AttentionLayer': AttentionLayer,
                        'LabelSmoothingLoss': LabelSmoothingLoss,
                        'WarmUpCosineDecay': WarmUpCosineDecay
                    },
                    compile=False
                )
                print(f"✓ Successfully loaded pretrained model")

                # Try to read epoch from checkpoint metadata
                checkpoint_epoch = 0
                metadata_path = model_path.replace('.keras', '_metadata.json')
                if os.path.exists(metadata_path):
                    try:
                        with open(metadata_path, 'r') as f:
                            metadata = json.load(f)
                            checkpoint_epoch = metadata.get('last_epoch', 0)
                            print(f"✓ Model metadata indicates training up to epoch {checkpoint_epoch}")
                    except:
                        pass

                return model, model_path, checkpoint_epoch
            except Exception as e:
                print(f"⚠ Failed to load {model_path}: {e}")
                continue

    return None, None, 0

# Check for existing training log
log_file = os.path.join(output_dir, 'training_log.csv')
initial_epoch = 0

print("\n" + "="*80)
print("CHECKING FOR PREVIOUS TRAINING SESSION")
print("="*80)

initial_epoch = get_last_completed_epoch(log_file)

if initial_epoch > 0:
    print(f"\n🔄 RESUMING TRAINING from epoch {initial_epoch}")
    if initial_epoch >= EPOCHS:
        print(f"\n⚠ WARNING: Training already completed!")
        print(f"  Completed: {initial_epoch} epochs")
        print(f"  Target: {EPOCHS} epochs")
        print(f"\n  Options:")
        print(f"  1. Increase EPOCHS to continue training")
        print(f"  2. Set USE_TRANSFER_LEARNING=False to restart")
        print(f"  3. Delete/rename the training log to start fresh")
else:
    print("\n✓ No previous training found. Starting from scratch.")

# Try to load pretrained model if transfer learning is enabled
pretrained_model = None
model_loaded_from = None
model_checkpoint_epoch = 0

if USE_TRANSFER_LEARNING or initial_epoch > 0:
    print("\n" + "="*80)
    print("LOADING MODEL FOR RESUME/TRANSFER LEARNING")
    print("="*80)

    # Define LabelSmoothingLoss before loading model
    class LabelSmoothingLoss(keras.losses.Loss):
        def __init__(self, smoothing=0.1, **kwargs):
            super().__init__(**kwargs)
            self.smoothing = smoothing

        def call(self, y_true, y_pred):
            y_true = tf.cast(y_true, tf.int32)
            y_true_one_hot = tf.one_hot(y_true, depth=NUM_CLASSES)
            y_true_smooth = y_true_one_hot * (1 - self.smoothing) + self.smoothing / NUM_CLASSES
            loss = -tf.reduce_sum(y_true_smooth * tf.math.log(y_pred + 1e-7), axis=-1)
            return tf.reduce_mean(loss)

    pretrained_model, model_loaded_from, model_checkpoint_epoch = load_pretrained_model()

    if pretrained_model is None:
        if initial_epoch > 0:
            print("\n❌ ERROR: Training log indicates previous training, but no model found!")
            print("   Cannot resume without saved model. Please:")
            print("   1. Restore the model file, or")
            print("   2. Delete the training_log.csv to start fresh")
            raise FileNotFoundError("Cannot resume training without saved model")
        else:
            print("\n⚠ No pretrained model found. Training from scratch...")
            USE_TRANSFER_LEARNING = False
    else:
        print(f"✓ Model loaded from: {model_loaded_from}")

        # Use the maximum of CSV epoch and model metadata epoch
        if model_checkpoint_epoch > initial_epoch:
            print(f"⚠ Model metadata shows epoch {model_checkpoint_epoch}, but CSV shows {initial_epoch}")
            print(f"  Using model metadata epoch: {model_checkpoint_epoch}")
            initial_epoch = model_checkpoint_epoch
else:
    # Define LabelSmoothingLoss for new training
    class LabelSmoothingLoss(keras.losses.Loss):
        def __init__(self, smoothing=0.1, **kwargs):
            super().__init__(**kwargs)
            self.smoothing = smoothing

        def call(self, y_true, y_pred):
            y_true = tf.cast(y_true, tf.int32)
            y_true_one_hot = tf.one_hot(y_true, depth=NUM_CLASSES)
            y_true_smooth = y_true_one_hot * (1 - self.smoothing) + self.smoothing / NUM_CLASSES
            loss = -tf.reduce_sum(y_true_smooth * tf.math.log(y_pred + 1e-7), axis=-1)
            return tf.reduce_mean(loss)

# ============================================================================
# BUILD MODEL
# ============================================================================

# Build or use model
if (USE_TRANSFER_LEARNING or initial_epoch > 0) and pretrained_model is not None:
    print("\n🔨 Using pretrained model...")
    model = pretrained_model

    # Optionally freeze CNN layers
    if FREEZE_CNN_LAYERS:
        print("\n🔒 Freezing CNN layers...")
        frozen_count = 0
        for layer in model.layers:
            if 'cnn_' in layer.name:
                layer.trainable = False
                frozen_count += 1
        print(f"✓ Frozen {frozen_count} CNN layers")
        print("✓ Only LSTM, Attention, and Dense layers will be trained")
    else:
        print("✓ All layers will be fine-tuned")

    model.summary()
else:
    print("\n🔨 Building new model from scratch...")
    model = build_adaptive_ser_model((MAX_TIME_STEPS, N_MELS, N_CHANNELS), NUM_CLASSES)
    model.summary()

# ============================================================================
# ADAPTIVE COMPILATION (SMART RESUME PATCH)
# ============================================================================

# Calculate steps
steps_per_epoch = len(train_paths) // BATCH_SIZE
total_steps = steps_per_epoch * EPOCHS
remaining_epochs = EPOCHS - initial_epoch
remaining_steps = remaining_epochs * steps_per_epoch

print(f"\n📊 Training Configuration:")
print(f"  • Total steps: {total_steps}")
print(f"  • Remaining steps: {remaining_steps}")
print(f"  • Starting from epoch: {initial_epoch}")

# SMART RESUME: Adjust learning rate based on resume point
if initial_epoch > 10:
    # Late resume - use custom schedule that continues from current position
    print(f"\n⏩ RESUMING LATE (Epoch {initial_epoch}). USING ADAPTIVE RESUME LR.")

    warmup_steps = int(total_steps * WARMUP_RATIO)

    # Adjust learning rate for transfer learning
    effective_lr = TARGET_LR
    if USE_TRANSFER_LEARNING and FREEZE_CNN_LAYERS:
        effective_lr = TARGET_LR * 0.1
        print(f"  • Reduced LR for frozen layers: {effective_lr}")

    # Create a custom schedule that accounts for the initial_epoch offset
    class ResumedCosineDecay(keras.optimizers.schedules.LearningRateSchedule):
        def __init__(self, target_lr, total_steps, warmup_steps, resume_step):
            super().__init__()
            self.target_lr = target_lr
            self.total_steps = total_steps
            self.warmup_steps = warmup_steps
            self.resume_step = resume_step
            self.decay_steps = total_steps - warmup_steps

        def __call__(self, step):
            # Add the resume offset to get the absolute step in the training schedule
            absolute_step = step + self.resume_step
            absolute_step = tf.cast(absolute_step, tf.float32)

            warmup_steps = tf.cast(self.warmup_steps, tf.float32)
            decay_steps = tf.cast(self.decay_steps, tf.float32)

            # Calculate position in cosine decay (we're past warmup)
            decay_step = absolute_step - warmup_steps
            cosine_decay = 0.5 * (1 + tf.cos(
                tf.constant(np.pi, dtype=tf.float32) * decay_step / decay_steps
            ))
            lr = self.target_lr * cosine_decay

            return lr

        def get_config(self):
            return {
                'target_lr': self.target_lr,
                'total_steps': self.total_steps,
                'warmup_steps': self.warmup_steps,
                'resume_step': self.resume_step
            }

    resume_step = initial_epoch * steps_per_epoch
    lr_schedule = ResumedCosineDecay(
        target_lr=effective_lr,
        total_steps=total_steps,
        warmup_steps=warmup_steps,
        resume_step=resume_step
    )

    # Calculate and display current LR
    current_step = initial_epoch * steps_per_epoch
    decay_step = current_step - warmup_steps
    decay_steps = total_steps - warmup_steps
    cosine_decay = 0.5 * (1 + np.cos(np.pi * decay_step / decay_steps))
    current_lr = effective_lr * cosine_decay

    print(f"   - Current LR at epoch {initial_epoch}: {current_lr:.6f}")
    print(f"   - Will continue cosine decay to epoch {EPOCHS}")

elif initial_epoch > 0:
    # Early resume - continue with warmup if needed
    print(f"\n🔄 RESUMING EARLY (Epoch {initial_epoch}). CONTINUING WARMUP.")
    warmup_steps = int(total_steps * WARMUP_RATIO)

    # Adjust learning rate for transfer learning
    effective_lr = TARGET_LR
    if USE_TRANSFER_LEARNING and FREEZE_CNN_LAYERS:
        effective_lr = TARGET_LR * 0.1
        print(f"  • Reduced LR for frozen layers: {effective_lr}")

    lr_schedule = WarmUpCosineDecay(
        target_lr=effective_lr,
        total_steps=total_steps,
        warmup_steps=warmup_steps
    )
    print(f"  • Warmup steps: {warmup_steps} (first {WARMUP_RATIO*100:.0f}%)")
    print(f"  • Target LR: {effective_lr}")

else:
    # Fresh start - use full warmup
    print(f"\n🟢 STARTING FRESH. USING FULL WARMUP.")
    warmup_steps = int(total_steps * WARMUP_RATIO)

    # Adjust learning rate for transfer learning
    effective_lr = TARGET_LR
    if USE_TRANSFER_LEARNING and FREEZE_CNN_LAYERS:
        effective_lr = TARGET_LR * 0.1
        print(f"  • Reduced LR for frozen layers: {effective_lr}")

    lr_schedule = WarmUpCosineDecay(
        target_lr=effective_lr,
        total_steps=total_steps,
        warmup_steps=warmup_steps
    )
    print(f"  • Warmup steps: {warmup_steps} (first {WARMUP_RATIO*100:.0f}%)")
    print(f"  • Target LR: {effective_lr}")

# Compile model with adaptive learning schedule
loss_fn = LabelSmoothingLoss(smoothing=LABEL_SMOOTHING)
optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)

model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

print("\n✓ Model compiled with Adaptive Learning Schedule (SMART RESUME)")
print(f"  • 3-Channel Input: Mel + Delta + Delta-Delta")
print(f"  • Attention Mechanism: ENABLED")
print(f"  • Rectangular Kernels: (5,3), (3,5), (3,3)")
if initial_epoch > 0:
    print(f"  • Resuming from epoch: {initial_epoch}")
    print(f"  • Learning rate adjusted for resume point")
elif USE_TRANSFER_LEARNING:
    print(f"  • Transfer Learning: ENABLED (loaded from: {model_loaded_from})")
    print(f"  • CNN Layers Frozen: {FREEZE_CNN_LAYERS}")

# ============================================================================
# CUSTOM CALLBACK FOR LEARNING RATE LOGGING
# ============================================================================

class LearningRateLogger(keras.callbacks.Callback):
    """Log learning rate at each epoch"""
    def __init__(self):
        super().__init__()
        self.lrs = []

    def on_epoch_begin(self, epoch, logs=None):
        # Get current learning rate from the schedule
        if hasattr(self.model.optimizer.learning_rate, '__call__'):
            # It's a schedule, call it with current step
            lr = self.model.optimizer.learning_rate(self.model.optimizer.iterations)
            lr = float(tf.keras.backend.get_value(lr))
        else:
            # It's a fixed value
            lr = float(tf.keras.backend.get_value(self.model.optimizer.learning_rate))

        self.lrs.append(lr)
        print(f"\nEpoch {epoch+1} - Learning Rate: {lr:.6f}")

lr_logger = LearningRateLogger()

# ============================================================================
# CALLBACKS
# ============================================================================

# Determine CSV logger mode
csv_append = initial_epoch > 0

callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath=os.path.join(checkpoint_dir, 'best_adaptive_model.keras'),
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    ),
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=20,
        restore_best_weights=True,
        verbose=1,
        min_delta=0.005
    ),
    keras.callbacks.CSVLogger(log_file, append=csv_append),  # Append if resuming
    keras.callbacks.TerminateOnNaN(),
    lr_logger
]

# Add custom callback to save metadata with checkpoints
class ModelMetadataSaver(keras.callbacks.Callback):
    """Save epoch metadata when model checkpoint is saved"""
    def on_epoch_end(self, epoch, logs=None):
        # Save metadata for the best model
        metadata_path = os.path.join(checkpoint_dir, 'best_adaptive_model_metadata.json')
        metadata = {
            'last_epoch': epoch,
            'val_accuracy': float(logs.get('val_accuracy', 0)),
            'val_loss': float(logs.get('val_loss', 0)),
            'timestamp': pd.Timestamp.now().isoformat()
        }
        with open(metadata_path, 'w') as f:
            json.dump(metadata, f, indent=2)

callbacks.append(ModelMetadataSaver())

if csv_append:
    print(f"\n✓ CSV Logger will APPEND to: {log_file}")
else:
    print(f"\n✓ CSV Logger will CREATE new file: {log_file}")

print(f"✓ Model metadata will be saved to Drive after each epoch")

# ============================================================================
# TRAIN
# ============================================================================

print("\n" + "="*80)
if initial_epoch > 0:
    print(f"RESUMING TRAINING FROM EPOCH {initial_epoch} (3-CHANNEL + ATTENTION)")
elif USE_TRANSFER_LEARNING:
    print("STARTING TRANSFER LEARNING (3-CHANNEL + ATTENTION)")
else:
    print("STARTING TRAINING FROM SCRATCH (3-CHANNEL + ATTENTION)")
print("="*80 + "\n")

# Check if training is already complete
if initial_epoch >= EPOCHS:
    print(f"\n⚠ Training already complete! ({initial_epoch}/{EPOCHS} epochs)")
    print("Skipping training. Loading final model for evaluation...")

    # Load the best model for evaluation
    best_model_path = os.path.join(checkpoint_dir, 'best_adaptive_model.keras')
    if os.path.exists(best_model_path):
        model = keras.models.load_model(
            best_model_path,
            custom_objects={
                'AttentionLayer': AttentionLayer,
                'LabelSmoothingLoss': LabelSmoothingLoss,
                'WarmUpCosineDecay': WarmUpCosineDecay
            }
        )
        print(f"✓ Loaded best model from {best_model_path}")

    history = None  # No new training history
else:
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=EPOCHS,
        initial_epoch=initial_epoch,  # START FROM LAST COMPLETED EPOCH
        callbacks=callbacks,
        verbose=1
    )

    print("\n✓ TRAINING COMPLETED")

# ============================================================================
# EVALUATE
# ============================================================================

print("\nEvaluating on test set...")
test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print(f"\n✓ Test Accuracy: {test_acc*100:.2f}%")
print(f"✓ Test Loss: {test_loss:.4f}")

# Predictions
y_true, y_pred = [], []
for mel_specs, labels in test_ds:
    preds = model.predict(mel_specs, verbose=0)
    y_pred.extend(np.argmax(preds, axis=1))
    y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# ============================================================================
# VISUALIZATIONS
# ============================================================================

# Training History + Learning Rate
fig = plt.figure(figsize=(18, 5))

# Load complete training history from CSV for plotting
if os.path.exists(log_file):
    try:
        history_df = pd.read_csv(log_file)

        # Accuracy
        ax1 = plt.subplot(1, 3, 1)
        ax1.plot(history_df['epoch'], history_df['accuracy'], label='Train', linewidth=2)
        ax1.plot(history_df['epoch'], history_df['val_accuracy'], label='Validation', linewidth=2)

        if initial_epoch > 0:
            ax1.axvline(x=initial_epoch-1, color='red', linestyle='--',
                      label=f'Resumed from epoch {initial_epoch}', alpha=0.7)

        warmup_epoch = (int(total_steps * WARMUP_RATIO)) // steps_per_epoch
        ax1.axvline(x=warmup_epoch, color='orange', linestyle='--',
                    label=f'Warmup End (Epoch {warmup_epoch})', alpha=0.7)

        title_suffix = " (Resumed)" if initial_epoch > 0 else " (Transfer Learning)" if USE_TRANSFER_LEARNING else ""
        ax1.set_title(f'Accuracy (3-Channel + Attention){title_suffix}', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Accuracy')
        ax1.legend()
        ax1.grid(alpha=0.3)

        # Loss
        ax2 = plt.subplot(1, 3, 2)
        ax2.plot(history_df['epoch'], history_df['loss'], label='Train', linewidth=2)
        ax2.plot(history_df['epoch'], history_df['val_loss'], label='Validation', linewidth=2)

        if initial_epoch > 0:
            ax2.axvline(x=initial_epoch-1, color='red', linestyle='--',
                      label=f'Resumed from epoch {initial_epoch}', alpha=0.7)

        ax2.axvline(x=warmup_epoch, color='orange', linestyle='--',
                    label=f'Warmup End', alpha=0.7)

        ax2.set_title(f'Loss (Adaptive Learning){title_suffix}', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.legend()
        ax2.grid(alpha=0.3)

    except Exception as e:
        print(f"⚠ Could not plot training history from CSV: {e}")
elif history is not None:
    # Fallback to history object if CSV not available
    warmup_epoch = (int(total_steps * WARMUP_RATIO)) // steps_per_epoch

    # Accuracy
    ax1 = plt.subplot(1, 3, 1)
    ax1.plot(history.history['accuracy'], label='Train', linewidth=2)
    ax1.plot(history.history['val_accuracy'], label='Validation', linewidth=2)
    ax1.axvline(x=warmup_epoch, color='orange', linestyle='--',
                label=f'Warmup End (Epoch {warmup_epoch})', alpha=0.7)

    title_suffix = " (Transfer Learning)" if USE_TRANSFER_LEARNING else ""
    ax1.set_title(f'Accuracy (3-Channel + Attention){title_suffix}', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(alpha=0.3)

    # Loss
    ax2 = plt.subplot(1, 3, 2)
    ax2.plot(history.history['loss'], label='Train', linewidth=2)
    ax2.plot(history.history['val_loss'], label='Validation', linewidth=2)
    ax2.axvline(x=warmup_epoch, color='orange', linestyle='--',
                label=f'Warmup End', alpha=0.7)
    ax2.set_title(f'Loss (Adaptive Learning){title_suffix}', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(alpha=0.3)

# Learning Rate (always from logger)
ax3 = plt.subplot(1, 3, 3)
if len(lr_logger.lrs) > 0:
    epochs_range = range(initial_epoch + 1, initial_epoch + 1 + len(lr_logger.lrs))
    ax3.plot(epochs_range, lr_logger.lrs, linewidth=2, color='green')

    if initial_epoch <= 10:
        warmup_epoch = (int(total_steps * WARMUP_RATIO)) // steps_per_epoch
        ax3.axvline(x=warmup_epoch, color='orange', linestyle='--',
                    label=f'Warmup End', alpha=0.7)

    ax3.set_title('Learning Rate Schedule', fontsize=14, fontweight='bold')
    ax3.set_xlabel('Epoch')
    ax3.set_ylabel('Learning Rate')
    ax3.legend()
    ax3.grid(alpha=0.3)
    ax3.set_yscale('log')

plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'training_history_adaptive.png'), dpi=300)
plt.show()

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=EMOTION_CLASSES, yticklabels=EMOTION_CLASSES)
axes[0].set_title('Confusion Matrix (Counts)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('True')

sns.heatmap(cm_norm, annot=True, fmt='.2%', cmap='Blues', ax=axes[1],
            xticklabels=EMOTION_CLASSES, yticklabels=EMOTION_CLASSES)
axes[1].set_title('Confusion Matrix (Normalized)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('True')

plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'confusion_matrix_adaptive.png'), dpi=300)
plt.show()

# ============================================================================
# CLASSIFICATION REPORT
# ============================================================================

print("\n" + "="*80)
print("CLASSIFICATION REPORT - ADAPTIVE LEARNING MODEL")
print("="*80 + "\n")

report = classification_report(y_true, y_pred, target_names=EMOTION_CLASSES, digits=4)
print(report)

with open(os.path.join(output_dir, 'classification_report_adaptive.txt'), 'w') as f:
    f.write("Speech Emotion Recognition - Adaptive Learning Model\n")
    f.write("="*80 + "\n\n")
    f.write("ARCHITECTURE:\n")
    f.write(f"  • 3-Channel Input: Mel + Delta + Delta-Delta ({N_MELS} × {N_CHANNELS})\n")
    f.write(f"  • Rectangular CNN Kernels: (5,3), (3,5), (3,3)\n")
    f.write(f"  • Attention Mechanism: ENABLED (silence handling)\n")
    f.write(f"  • Adaptive Learning: WarmUp + Cosine Decay\n")
    f.write(f"  • Target LR: {TARGET_LR}\n")
    if initial_epoch > 0:
        f.write(f"  • Training Resumed from Epoch: {initial_epoch}\n")
    elif USE_TRANSFER_LEARNING:
        f.write(f"  • Transfer Learning: ENABLED (from: {model_loaded_from})\n")
        f.write(f"  • CNN Layers Frozen: {FREEZE_CNN_LAYERS}\n")
    f.write("="*80 + "\n\n")
    f.write(report)
    f.write(f"\n\nTest Accuracy: {test_acc*100:.2f}%")
    f.write(f"\nTest Loss: {test_loss:.4f}")

# Save final model
final_model_path = os.path.join(output_dir, 'final_adaptive_model.keras')
model.save(final_model_path)
print(f"\n✓ Final model saved: {final_model_path}")

# ============================================================================
# SUMMARY
# ============================================================================

print("\n" + "="*80)
print("ADAPTIVE LEARNING STRATEGY - 4 PILLARS IMPLEMENTED")
print("="*80)
print("\n✅ PILLAR 1: 3-Channel Input (Physics-Aware)")
print(f"   → Mel-Spectrogram + Delta + Delta-Delta")
print(f"   → Shape: ({MAX_TIME_STEPS}, {N_MELS}, {N_CHANNELS})")

print("\n✅ PILLAR 2: Attention Mechanism (Silence Handling)")
print(f"   → Custom AttentionLayer implemented")
print(f"   → Learns to ignore silence/padding at clip boundaries")

print("\n✅ PILLAR 3: Rectangular CNN Kernels")
print(f"   → Block 1: (5,3) - Time focus")
print(f"   → Block 2: (3,5) - Frequency focus")
print(f"   → Block 3: (3,3) - Texture focus")

print("\n✅ PILLAR 4: Adaptive Learning (WarmUp + Cosine Decay)")
print(f"   → Smart Resume: Learning rate adjusted for epoch {initial_epoch}")
print(f"   → Target LR: {TARGET_LR}")
print(f"   → Smooth cosine decay")

print(f"\n📊 FINAL RESULTS:")
print(f"   → Test Accuracy: {test_acc*100:.2f}%")
print(f"   → Test Loss: {test_loss:.4f}")
print("="*80 + "\n")
```
